```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv("Cryogenic Storage Tank Sales data.xlsx - Cryogenic sales data.csv")
df
```

```python
df.rename(columns={"Sales in Thousands":"sales"},inplace=True)
df.head()
```

```python
df1 = df[["sales",'month']]
df1.head()
```

```python
sales_by_month = df1.groupby('month')['sales'].sum()
sales_by_month
```

```python
sales_by_month_df = sales_by_month.reset_index()

# Rename the columns
sales_by_month_df.columns = ['month', 'sales']

print(sales_by_month_df)
```

```python
df2 = sales_by_month_df.reset_index()['sales']
df2
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from matplotlib import pyplot as plt
from pandas import read_csv
from pandas import datetime 
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error,mean_absolute_error
from math import sqrt
model1 = ARIMA(df2.values, order=(5,1,0))
model_fit1 = model1.fit()
output1= model_fit1.forecast(steps=24)
output1
```

```python
train_size = int(len(sales_by_month_df["sales"]) * 0.70)
train_data, test_data = sales_by_month_df["sales"][:train_size], sales_by_month_df["sales"][train_size:]
model1 = ARIMA(train_data, order=(5,1,0))
model_fit1 = model1.fit()
output1= model_fit1.forecast(steps=len(test_data))
output1
```
